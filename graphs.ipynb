```python
import matplotlib.pyplot as plt 
x = [2,6]
y = [4,9]
x1 = [8,3]
y1 = [4,7]
plt.plot(x,y,label='river',c='blue')
plt.plot(x1,y1,label='grass',c='green')
plt.xlabel("Time")
plt.ylabel("speed")
plt.legend()
plt.grid(color='grey')
plt.show()
```

```python
x=[100,60,34,90]
y=["virat","rahul","sachin","yuvi"]
plt.ylabel("score")
plt.xlabel("P.Name")
plt.bar(y,x,label="overs")
plt.legend()
plt.grid(color='y')
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [2,6,10]
y = [4,9,7]
x1 = [3,5,7]
y1 = [5,8,10]
plt.scatter(x,y,label='river',c='blue',marker='x',s=120)
plt.scatter(x1,y1,label='grass',c='green',s=120)
plt.xlabel("Time")
plt.ylabel("speed")
plt.legend()
plt.grid(color='y')
plt.show()
```

```python
import matplotlib.pyplot as plt 
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```
